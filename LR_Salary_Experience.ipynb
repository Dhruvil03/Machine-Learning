#Performing Linear Regression on CSV File

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
def ss(exp_train,sala_train):
  reg = LinearRegression().fit(exp_train,sala_train)
  return reg


df = pd.read_csv('Salary_Data.csv')
ex = df['YearsExperience']
sal = df['Salary']

ex = np.reshape(np.array(ex),(len(ex),1))
sal = np.reshape(np.array(sal),(len(sal),1))

ex_train,ex_test,sal_train,sal_test = train_test_split(ex,sal)

reg1 = ss(ex_train,sal_train)

print(reg1.coef_)
print(reg1.intercept_)
print(reg1.predict([[6]])) #here Years of Experience =6

print(reg1.score(ex_train,sal_train))
print(reg1.score(ex_test,sal_test))
```

#Testing the accuracy

```python
exp = df['YearsExperience']

exp = np.reshape(np.array(exp),(len(exp),1))
print(exp)
y_true = df['Salary']
y_pred = reg1.predict(exp)
print(y_true)
print(y_pred)
r2_score(y_pred,y_true)
```

#Ploting A Graph for linear Regression

```python
plt.figure(figsize = (12,10))
sns.regplot(x = ex_train,y=sal_train,scatter = True, color="r",marker = "*")
plt.xlabel("Experience_Train")
plt.ylabel("Salary_train")
plt.title("Regression Plot")
```

#Ploting Graph for Train and Test Data

```python
plt.figure(figsize = (12,10))
plt.scatter(x=ex_train,y=sal_train,color='cyan',label="Train_Data")
plt.scatter(x=ex_test,y=sal_test,color='r',label="Test_Data")
plt.plot(ex_test,reg1.predict(ex_test)) #Ploting A Line Based on test data and predict data of experience
plt.xlabel("Years of Experiece")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
