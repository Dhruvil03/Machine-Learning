##**Breast Cancer Detection using Logistic Regression & KNN Classification**

```python
#importing libraries
import numpy as np
import sklearn.datasets
```

```python
#getting the dataset
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
# 0 = malignant
# 1 = benign
```

```python
X = breast_cancer.data
Y = breast_cancer.target
```

```python
print(X)   #data
print(Y)   #equivalent labels
```

```python
print(X.shape, Y.shape)
```

##Importing data to Pandas Data Frame

```python
import pandas as pd
```

```python
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
data['class'] = breast_cancer.target
```

```python
data.describe()
```

##Train and Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
print(Y.shape,y_train.shape,y_test.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1) #Percentage of data held for testing i.e. 10% will be taken for testing
```

```python
print(Y.shape,y_train.shape,y_test.shape)
```

```python
print(Y.mean(),y_train.mean(),y_test.mean())
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,stratify=Y)
#stratify -> to overcome train_test_split's biased random sampling problem i.e overfitting and underfitting
```

```python
print(Y.mean(),y_train.mean(),y_test.mean())
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
#It means one random_state value has a fixed dataset.
#It means every time we run code with random_state value 1, it will produce the same splitting datasets.
```

```python
print(X.mean(),x_train.mean(),x_test.mean())
```

```python
print(x_train)
```

##**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
#Training Model on Training Data
classifier.fit(x_train,y_train)
```

##Evaluation of model

```python
from sklearn.metrics import accuracy_score
```

```python
prediction_on_training_data = classifier.predict(x_train)
accuracy_train = accuracy_score(y_train,prediction_on_training_data)
print("Accuracy on training data = ", accuracy_train)
```

```python
#prediction on Test data
prediction_on_testing_data = classifier.predict(x_test)
accuracy_test = accuracy_score(y_test,prediction_on_testing_data)
print("Accuracy on testing data = ", accuracy_test)
```
